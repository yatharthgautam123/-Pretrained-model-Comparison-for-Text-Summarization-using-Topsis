```python
import matplotlib.pyplot as plt
import pandas as pd
topsis_results = pd.read_csv("/kaggle/input/topsis-text-summarization-results/topsis_results.csv")

# Visualize TOPSIS scores
plt.figure(figsize=(10, 5))
plt.bar(topsis_results["Model"], topsis_results["TOPSIS Score"], color='blue')
plt.title('TOPSIS Scores for Each Model')
plt.xlabel('Model')
plt.ylabel('TOPSIS Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize TOPSIS ranks
plt.figure(figsize=(10, 5))
plt.bar(topsis_results["Model"], topsis_results["Rank"], color='green')
plt.title('TOPSIS Ranks for Each Model')
plt.xlabel('Model')
plt.ylabel('Rank')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Visualize comparison of evaluation metrics
metric_names = ["BLEU Score", "Semantic Coherence", "Factual Accuracy", "Content Coverage"]

for metric in metric_names:
    plt.figure(figsize=(10, 5))
    plt.bar(topsis_results["Model"], topsis_results[metric], color='orange')
    plt.title(f'{metric} Across Models')
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
# Calculate the correlation matrix
correlation_matrix = topsis_results.iloc[:, 1:].corr()

# Visualize the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Evaluation Metrics')
plt.show()
```

```python
# Visualize pairwise scatter plots
sns.pairplot(topsis_results.iloc[:, 1:])
plt.suptitle('Pairwise Scatter Plots of Evaluation Metrics', y=1.02)
plt.show()
```
