```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have a DataFrame with your model comparison data
# Replace the following line with your actual DataFrame
data = {
    'Model': ['BERTSumExt', 'GPT-3', 'T5', 'XLNet', 'BART', 'Pegasus'],
    'Rouge Scores': [0.75, 0.82, 0.78, 0.76, 0.80, 0.79],
    'Length of Summary': [130, 150, 140, 135, 145, 138],
    'Training Time': [9, 12, 10, 11, 8, 13],
    'Rank': [3, 1, 2, 4, 5, 6]
}

df = pd.DataFrame(data)

# Sorting the DataFrame by Rank for better visualization
df = df.sort_values(by='Rank')

# Create a bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
bar_positions = range(len(df['Model']))

for i, metric in enumerate(['Rouge Scores', 'Length of Summary', 'Training Time']):
    ax.bar(
        [pos + i * bar_width for pos in bar_positions],
        df[metric],
        width=bar_width,
        label=metric
    )

ax.set_xticks([pos + bar_width for pos in bar_positions])
ax.set_xticklabels(df['Model'])
ax.legend()
ax.set_xlabel('Model')
ax.set_title('Model Comparison Metrics')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming you have a DataFrame with your model comparison data
# Replace the following line with your actual DataFrame
data = {
    'Model': ['BERTSumExt', 'GPT-3', 'T5', 'XLNet', 'BART', 'Pegasus'],
    'Rouge Scores': [0.75, 0.82, 0.78, 0.76, 0.80, 0.79],
    'Length of Summary': [130, 150, 140, 135, 145, 138],
    'Training Time': [9, 12, 10, 11, 8, 13],
    'Rank': [3, 1, 2, 4, 5, 6]
}

df = pd.DataFrame(data)

# Sorting the DataFrame by Rank for better visualization
df = df.sort_values(by='Rank')

# Set a color palette for better distinction
sns.set_palette("viridis")

# Create a bar chart
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.2
bar_positions = range(len(df['Model']))

# Plot each metric separately for better clarity
for i, metric in enumerate(['Rouge Scores', 'Length of Summary', 'Training Time']):
    ax.bar(
        [pos + i * bar_width for pos in bar_positions],
        df[metric],
        width=bar_width,
        label=metric
    )

ax.set_xticks([pos + bar_width for pos in bar_positions])
ax.set_xticklabels(df['Model'])
ax.legend()
ax.set_xlabel('Model')
ax.set_title('Model Comparison Metrics')

# Display the values on top of each bar
for i, metric in enumerate(['Rouge Scores', 'Length of Summary', 'Training Time']):
    for j, value in enumerate(df[metric]):
        ax.text(j + i * bar_width, value + 0.02, f'{value}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
